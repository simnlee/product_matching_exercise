# Create labeled ground truth data

```python
import pandas as pd
requests_df = pd.read_csv('documents/Request-Response 1/request_1.csv')
requests_df.head()
```

```python
responses_df = pd.read_csv('documents/Request-Response 1/response_1.csv')
responses_df.head()
```

```python
productdb_df = pd.read_csv('documents/ProductDB_clean.csv')
```

```python
productdb_df
```

```python
def get_product_name(product_no: int):
    row = productdb_df[productdb_df['PRODUCT_NO'] == product_no]
    return row['NAME'].iloc[0] + ' ' + str(row['ALIAS'].iloc[0])
```

```python
def get_customer_name(quantity: int):
    for index, row in requests_df.iterrows():
        if round(row['Quantity']) == quantity:
            return row['Description']
```

```python
get_customer_name(35289)
```

```python
product_nos = []
customer_names = []
for index, row in responses_df.iterrows():
    length_str = row['Length (FT)'].replace(',', '')
    length = round(float(length_str))
    customer_name = get_customer_name(length)
    product_nos.append(row['PRODUCT'])
    customer_names.append(customer_name)
```

```python
# create df out of product_nos and customer_names
labels_df = pd.DataFrame(list(zip(customer_names, product_nos)), columns=['Customer Name', 'Product No'])
labels_df
```

```python
manufacturer_names = []
for index, row in labels_df.iterrows():
    product_no = row['Product No']
    manufacturer_name = get_product_name(product_no)
    manufacturer_names.append(manufacturer_name)
```

```python
manufacturer_names
```

```python
labels_df.to_csv('documents/Request-Response 1/labels.csv', index=False)
```

```python
# get subset of productdb_df where PRODUCT_NO is in product_nos
productdb_subset_df = productdb_df[productdb_df['PRODUCT_NO'].isin(product_nos)]
productdb_subset_df
```

```python
productdb_subset_df.to_csv('documents/productdb_subset.csv', index=False)
```
