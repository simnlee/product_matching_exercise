# Import product catalog CSV into Pandas

```python
import pandas as pd

column_names = ['PRODUCT_ID', 'PRODUCT_NO', 'NAME', 'ALIAS', 'STATUS', 'PRODUCT_LINE_CODE', 'PRODUCT_GROUP_CODE', 'PRODUCT_CODE_CODE', 'DROP', 'Product', 'SFIA']

catalog_raw_df = pd.read_csv('documents/ProductDB.csv', names=column_names)
```

```python
catalog_raw_df.shape
```

```python
catalog_raw_df.head()
```

```python
# remove last row of catalog_raw_df
catalog_raw_df = catalog_raw_df.drop(catalog_raw_df.index[-1])
# drop DROP column
catalog_clean_df = catalog_raw_df.drop(columns=['DROP'])
```
